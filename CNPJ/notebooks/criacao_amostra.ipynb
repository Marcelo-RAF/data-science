```python
import pandas as pd
import numpy as np
import zipfile
```

```python
zip_path = "../data/raw/Estabelecimentos0.zip"

with zipfile.ZipFile(zip_path) as z:
    file_name = z.namelist()[0]
    with z.open(file_name) as f:
        df = pd.read_csv(
            f,
            sep=";",
            encoding="latin1",
            header=None,
            dtype=str,
            low_memory=False
        )
```

```python
# Escrevendo CPF completo
df["CNPJ"] = df[0] + df[1] + df[2]

df_reduzido = df[["CNPJ", 10]].copy()
df_reduzido.columns = ["CNPJ", "data_abertura"]

# Convertendo o formato de data
df_reduzido["data_abertura"] = pd.to_datetime(
    df_reduzido["data_abertura"],
    format="%Y%m%d",
    errors="coerce"
)

#Se tiver data inválida, apagar
df_reduzido = df_reduzido.dropna(subset=["data_abertura"])

# Escolhendo amostras aleatórias
amostra = df_reduzido.sample(n=100, random_state=42).reset_index(drop=True)

#Apagando 20 delas
indices_apagar = np.random.choice(amostra.index, size=20, replace=False)
amostra.loc[indices_apagar, "data_abertura"] = pd.NaT

amostra.to_csv("cnpjs_amostra.csv", index=False)

print("Arquivo salvo com sucesso.")
```
